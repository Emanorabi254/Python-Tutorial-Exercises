To display the image, there are 2 options:

1. using cv2.imshow
2. using plt.imshow

```python
print(help(cv2.imshow))
```

```python
cv2.imshow('image',img) # display the image
cv2.waitKey(0)   # Wait a period before closing the image "note that the delay is in the milli second"
cv2.destroyAllWindows()  # close all opening windows after the waiting delay.
```

Here with cv2.imshow() --> we first put name of the window that the image will be displayed in , then pass the variable that decleared with cv2.imread , but in matplot dosen't need those , it is only need the variable.

```python
plt.imshow(img)
plt.show()
```

As shown the image hadn't displayed in its original colour , but cv2 didin't do this.
This happend because cv2 reads the image in RGB format, but matplotlib displays it in BGR format , so we have to covert it first.

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

To save image

we use cv2.imwrite('path of the image that will be in', the variable that contain the image you wanna to save)

```python
saved_img="./imgs/copyimg.jpg"
ret=cv2.imwrite(saved_img,img)
if ret:
    print("The image is saved successfully as {saved_img}")
else:
    print("Failed to save :()")    
```

Define as imshow function to plot the image below the cell rather than in another window as it's done with cv2.imshow

```python
def imshow(img):
    copied_img=img.copy()  # take a copy or original image
    copied_img=cv2.cvtColor(copied_img,cv2.COLOR_BGR2RGB)
    plt.imshow(copied_img)
    plt.show()
imshow(img)    
```

____Image Cropping____

As we say that the image consider an array, if we need to take part from the image "__Cropping__" , we should a part from the array.

```python
cropped_img=img[100:300,200:500]  # [height : width]
cropped_img=cv2.cvtColor(cropped_img,cv2.COLOR_BGR2RGB)
plt.imshow(cropped_img)
plt.show()
```

__Scaling__

```python
resized_shape=(500,500)
resized_img=cv2.resize(img,resized_shape)
```

```python
cv2.imshow('Scalingimg',resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

___Channels Splitting___

To get each channel alone , we use the function of split() ---> it will return 3 arrays blue, green, red each with height and width like the original image , but with one channel.

```python
blue, green, red=cv2.split(img)
```

```python
print(f"red channel shape {red.shape}")  # note that the shape will be as the orginal img but with one channel
imshow(red)  # gray scale image , "note that this function is what we do above."
```

```python
imshow(green) 
```

```python
imshow(blue) 
```

if you want to highlight any desired channel , you can use cv2.merge and put zeros in the other channel :)

```python
zeros_arr=np.zeros(red.shape,dtype='uint8')
red_image=cv2.merge([zeros_arr,zeros_arr,red])
imshow(red_image)
```

```python
blue_image=cv2.merge([blue,zeros_arr,zeros_arr])
imshow(blue_image)
```

```python
green_image=cv2.merge([zeros_arr,green,zeros_arr])
imshow(green_image)
```

```python
mixbr_image=cv2.merge([blue,zeros_arr,red])
imshow(mixbr_image)
```

```python
mixrg_image=cv2.merge([zeros_arr,green,red])
imshow(mixrg_image)
```

```python
mixbg_image=cv2.merge([blue,green,zeros_arr])
imshow(mixbg_image)
```
